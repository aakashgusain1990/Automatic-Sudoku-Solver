```
board = [
    [7,8,0,4,0,0,1,2,0],
    [6,0,0,0,7,5,0,0,9],
    [0,0,0,6,0,1,0,7,8],
    [0,0,7,0,4,0,2,6,0],
    [0,0,1,0,5,0,9,3,0],
    [9,0,4,0,6,0,0,0,5],
    [0,7,0,3,0,0,0,1,2],
    [1,2,0,0,0,7,4,0,0],
    [0,4,9,2,0,6,0,0,7]
]

def solve(bo):
  find = find_empty(bo)
  if not find:
    return True
  else:
    row,col = find

  for i in range(1,10):
    if isvalid(bo,i,find):
      bo[row][col]=i
      if(solve(bo)):
        return True
      bo[row][col]=0;
  return False

def isvalid(bo,num,pos):
  row,col = pos
  for i in range(len(bo[col])):
    if bo[row][i] == num:
      return False
  
  for i in range(len(bo[row])):
    if bo[i][col] == num:
      return False

  x=col//3
  y=row//3

  for i in range(3*x,3*x+3):
    for j in range(3*y,3*y):
      if bo[i][j] == num:
       return False
  
  return True
  
def find_empty(bo):
  for i in range(len(bo)):
    for j in range(len(bo[i])):
      if bo[i][j]==0:
        return (i,j)
  return None


solve(board)
for i in range(len(board)):
  for j in range(len(board[i])):
  
    print(board[i][j])
    
```
